```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("DiseasePrediction.csv")
dataset
```

```python
#convert categorical data into numeric data
from sklearn.preprocessing import LabelEncoder
```

```python
# Initialize LabelEncoder
disease=LabelEncoder()

dataset['Disease']=disease.fit_transform(dataset['Disease'])
dataset
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
symptoms_encoder = LabelEncoder()

# List of symptom columns
symptom_cols = ["Symptom_1", "Symptom_2", "Symptom_3", "Symptom_4", "Symptom_5",
                "Symptom_6", "Symptom_7", "Symptom_8", "Symptom_9", "Symptom_10",
                "Symptom_11", "Symptom_12", "Symptom_13", "Symptom_14", "Symptom_15",
                "Symptom_16", "Symptom_17"]

# Apply Label Encoding to each symptom column
for col in symptom_cols:
    dataset[col] = symptoms_encoder.fit_transform(dataset[col])
dataset
```

```python
#seperate x=independent(Symptom_1	Symptom_2	Symptom_3	Symptom_4	Symptom_5	Symptom_6	Symptom_7	Symptom_8	Symptom_9	Symptom_10
#Symptom_11	Symptom_12	Symptom_13	Symptom_14	Symptom_15	Symptom_16	Symptom_17) and y=dependent(Disease)
x=dataset[["Symptom_1",	"Symptom_2","Symptom_3","Symptom_4","Symptom_5","Symptom_6","Symptom_7","Symptom_8","Symptom_9","Symptom_10","Symptom_11","Symptom_12","Symptom_13","Symptom_14","Symptom_15","Symptom_16","Symptom_17"]].values
x
```

```python
#seperate y=dependent('Diseases')
y=dataset['Disease'].values
y
```

```python
#create train_test_split

from sklearn.model_selection import train_test_split

#test_train_data.... 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
#create Random forest classifier model

from sklearn.ensemble import RandomForestClassifier
```

```python
#create a instance of RandomForestClassifier

rfc=RandomForestClassifier()
```

```python
rfc.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
rfc.predict(x_test)
```

```python
rfc.predict([[32,350,5,1,33,39,6,1,2,2,1,1,3,4,6,7,1]])
```

```python
rfc.score(x_test,y_test)
```

```python
#now we can find out least important feature with rfc.feature_importances_

imp_features=pd.Series(rfc.feature_importances_,index=["Symptom_1",	"Symptom_2","Symptom_3","Symptom_4","Symptom_5","Symptom_6","Symptom_7","Symptom_8","Symptom_9","Symptom_10","Symptom_11","Symptom_12","Symptom_13","Symptom_14","Symptom_15","Symptom_16","Symptom_17"])
```
